# Exercícios de Pandas

### 1. Criando um DataFrame

```python
#Importando biblioteca pandas
import pandas as pd

#Dados
nomes = ['Berenice', 'Cacau', 'Madalena', 'Spike', 'Julio']
idades = [1, 4, 3, 5, 6]
cidades = ['Cabo Verde','São José','Poços de Caldas', 'Guaxupé', 'Botelhos']
sexos = ['F', 'F', 'F', 'M', 'M']
#Criando DataFrame
df = pd.DataFrame(list(zip(nomes, idades,cidades, sexos)), columns=['Nome', 'Idade','Cidade', 'Sexo'])
#Imprimindo DataFrame
print(df)
```

### 2. Acessando Dados

```python
#Acessa linha 2
print(df.loc[2])
#Acessa linha 3 na coluna nome
print(df.at[3, 'Nome'])
#Acessa célula 2,2
print(df.iat[2, 2])
```

### 3. Filtrando Dados

```python
#Filtrando Idade maior ou igual a 2
df.loc[df['Idade'] >= 2]
```

```python
#Filtrando idade menor ou igual a 2
df.loc[df['Idade'] <= 2]
```

### 4. Ordenando Dados

```python
#Ordenando dados por idade de forma decrescente
df.sort_values(by='Idade', ascending=False)
```

```python
#Ordenando dados por idade de forma crescente
df.sort_values(by='Idade', ascending=True)
```

### 5. Agregação de Dados

```python
#Agrupando dados pelo sexo a partir da média, mínimo e máximo de idade
df.groupby('Sexo').agg({
  'Idade': ['mean', 'min', 'max'],
  'Nome': 'count'
})
```

### 6. Adicionando uma Nova Coluna

```python
#Adicionando a coluna raça com os respectivos dados
df['Raça'] = ['Daschund', 'Cocker', 'Vira Lata', 'Labrador', 'Pinscher']
df
```

### 7. Removendo Colunas

```python
#Verifica se há a coluna Cidade e remove
if 'Cidade' in df.columns:
    df = df.drop(columns=['Cidade'])
#Atualização do DataFrame
print("\nDataFrame após remoção da coluna Cidade:")
print(df)
```

### 8. Salvando e Carregando Dados

```python
#Salva o arquivo em .CSV
df.to_csv('arquivo.csv', index=False)
df = pd.read_csv('arquivo.csv')
df
```

### 9. Consultas Condicionais

```python
#Consultando dados a partir do sexo feminino
df[df['Sexo'] == 'F']
```

```python
#Consultando dados a partir do sexo masculino
df[df['Sexo'] == 'M']
```

### 10. Contagem de Valores

```python
#Contando valores pela Idade
print("\nContagem de valores de cada Idade:")
print(df['Idade'].value_counts())
```

```python
# Contagem de valores pelo sexo
print("\nContagem de valores de cada Sexo:")
print(df['Sexo'].value_counts())
```

### 11. Substituição de Valores

```python
#Substituindo valor da coluna raça
df['Raça'] = df['Raça'].replace({
    'Vira Lata': 'SRD',
    'Daschund': 'Salsichinha',
})
#Atualização do Data Frame
print("\nDataFrame após substituição do nome da raça:")
print(df)
```

### 12. Resumo Estatístico

```python
#Resumi estatístico incluindo valores e categorias
print(df.describe(include='all'))
```

### 13. Preenchimento de Valores Ausentes

```python
# Preenchendo valores ausentes: números com 0, textos com 'Desconhecido'
df_preenchido = df.fillna({
    'Nome': 'Não informado',
    'Idade': 0,
    'Cidade': 'Não informada',
    'Sexo': 'Não declarado'
})

# Exibindo DataFrame preenchido
print("Valores preenchidos: ", df_preenchido)
```

### 14. Concatenação de DataFrames

```python
# Definindo o primeiro conjunto de dados
nomes1 = ['Berenice', 'Madalena', 'Julio']
idades1 = [1, 3, 6]
cidades1 = ['Cabo Verde','Poços de Caldas', 'Botelhos']
sexos1 = ['F', 'F', 'M']

# Segundo conjunto de dados
nomes2 = ['Cacau', 'Spike']
idades2 = [4, 5]
cidades2 = ['São José', 'Guaxupé']
sexos2 = ['F', 'M']

# Criando os dois DataFrames
df1 = pd.DataFrame(list(zip(nomes1, idades1, cidades1, sexos1)), columns=['Nome', 'Idade', 'Cidade', 'Sexo'])
df2 = pd.DataFrame(list(zip(nomes2, idades2, cidades2, sexos2)), columns=['Nome', 'Idade', 'Cidade', 'Sexo'])

# Concatenando verticalmente
df_concatenado = pd.concat([df1, df2], axis=0)

# Exibindo resultado
print(df_concatenado)
```

### 15. Mescle (Merge) de DataFrames

```python
# Primeiro DataFrame: Nome e Idade dos animais
df1 = pd.DataFrame({
    'Nome': ['Berenice', 'Cacau', 'Madalena'],
    'Idade': [1, 4, 3]
})

# Segundo DataFrame: Demais informações
df2 = pd.DataFrame({
    'Nome': ['Cacau', 'Madalena', 'Julio'],
    'Cidade': ['São José', 'Poços de Caldas', 'Botelhos']
})

# Realizando mesclagem com a coluna Nome
mesclado = pd.merge(df1, df2, on='Nome')

# Exibindo resultado
print(mesclado)
```

### 16. Agrupamento de Dados

```python
#Criando data frame
df = pd.DataFrame(list(zip(nomes, idades, cidades, sexos)), columns=['Nome', 'Idade','Cidade', 'Sexo'])

# Agrupando dados por Cidade e calculando média de Idade
agrupado = df.groupby('Cidade')['Idade'].mean()

# Exibindo resultado
print(agrupado)
```

### 17. Plotagem de Dados

```python
import matplotlib.pyplot as plt
# Contagem por Cidade
contagem_por_cidade = df['Cidade'].value_counts()

# Plotando gráfico de barras
contagem_por_cidade.plot(kind='bar')
plt.xlabel('Cidade')
plt.ylabel('Contagem')
plt.title('Número de Animais por Cidade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### 18. Filtros Avançados

```python
# Filtro avançado: Idade entre 3 e 5, Cidade igual a 'São José'
filtro_avancado = df[(df['Idade'] >= 3) & (df['Idade'] <= 5) & (df['Cidade'] == 'São José')]

print(filtro_avancado)
```

### 19. Pivotagem de Dados

```python
# Criando tabela pivot com média de idade por cidade
tabela_pivot = pd.pivot_table(df, values='Idade', index='Cidade', aggfunc='mean')

print(tabela_pivot)
```
